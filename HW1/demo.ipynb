```python
from PIL import Image

# Mở ảnh PNG
img = Image.open("500x1000.jpg").convert("RGB")
width, height = img.size
pixels = list(img.getdata())

# Ghi file P3
with open("500x1000.pnm", "w") as f:
    f.write("P3\n")
    f.write(f"{width} {height}\n")
    f.write("255\n")
    for r, g, b in pixels:
        f.write(f"{r} {g} {b} ")
```
